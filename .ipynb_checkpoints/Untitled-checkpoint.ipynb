```python
import numpy as np
print("파이썬 리스트를 이용한 연산")
a = [1, 2, 3]
b = [4, 5, 6]

print("a+b:", a+b)
try:
    print(a*b)
except TypeError:
    print("a*b 파이썬 리스트에 대해 a*b와 같은 연산을 할 수 없음")
print()
print("넘파이 배열을 사용한 연산")
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("a+b:",a+b)
print("a*b:",a*b)
```
